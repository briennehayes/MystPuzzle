```python
import networkx as nx
import re
import numpy as np
import pandas as pd
import random as rand
import matplotlib.pyplot as plt
from itertools import count, product
```

```python
graph_file = open('output.txt', 'r')
```

```python
with open('output.txt') as f:
    read_data = f.read()
```

```python
graph_file.close()
```

```python
graphs = read_data.split("\n")
```

```python
graphs2 = [line for line in graphs if line is not '']
```

```python
graph_names = [val for i,val in enumerate(graphs2) if i%3==0]
graph_attrs = [val for i,val in enumerate(graphs2) if i%3==1]
graph_edges = [val for i,val in enumerate(graphs2) if i%3==2]
```

```python
graph_df = pd.DataFrame({'Name' : graph_names, 'Attributes' : graph_attrs, 'Edges' : graph_edges})
```

```python
def graph_name_join(string):
    regex = r"Graph (\d+), order (\d+)"
    match = re.match(regex, string)
    output = "Graph " + match[2] + "-" + match[1]
    return(output)
```

```python
graph_df['Name'] = [graph_name_join(string) for string in graph_df['Name']]
```

```python
graph_df['Num_Nodes'] = [int(nums.split(' ')[0]) for nums in graph_df['Attributes']]
graph_df['Num_Edges'] = [int(nums.split(' ')[1]) for nums in graph_df['Attributes']]
```

```python
def build_edge_list(nums):
    L = [int(num) for num in nums.split(' ') if num is not '']
    it = iter(L)
    return list(zip(it, it))
```

```python
graph_df['Edges'] = [build_edge_list(nums) for nums in graph_df['Edges']]
```

```python
graph_df = graph_df.drop('Attributes', axis = 1)
```

```python
def build_graph(num_nodes, edge_list):
    G = nx.Graph()
    G.add_nodes_from(np.arange(num_nodes))
    nx.set_node_attributes(G, 0, name = 'color')
    G.add_edges_from(edge_list)
    return(G)
```

```python
graph_df['Graph_Objects'] = [build_graph(num_nodes, edge_list) for num_nodes, edge_list in zip(graph_df['Num_Nodes'], graph_df['Edges'])]
```

```python
def min_degree(G):
    return(min([tup[1] for tup in list(G.degree(np.arange(G.number_of_nodes())))]))
```

```python
graph_df['Min_Degree'] = [min_degree(G) for G in graph_df['Graph_Objects']]
```

```python
[row for row in graph_df if graph_df['Min_Degree'] is not 1]
```

```python
valid_graphs = graph_df[graph_df.Min_Degree > 1]
```

```python
invalid_graphs = graph_df[graph_df.Min_Degree == 1]
```
